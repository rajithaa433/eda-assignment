```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")

df.head()
```

Task 1: Distribution Analysis (Histograms)

```python
# Histogram with 5 bins
plt.figure(figsize=(6,4))
plt.hist(df['total_bill'], bins=5, color='skyblue', edgecolor='black')
plt.xlabel("Total Bill (USD)")
plt.ylabel("Frequency")
plt.title("Histogram of Total Bill (5 Bins)")
plt.show()

# Histogram with 20 bins
plt.figure(figsize=(6,4))
plt.hist(df['total_bill'], bins=20, color='orange', edgecolor='black')
plt.xlabel("Total Bill (USD)")
plt.ylabel("Frequency")
plt.title("Histogram of Total Bill (20 Bins)")
plt.show()
```

Distribution Shape:
The distribution of total_bill is right-skewed (positively skewed). Most bills are concentrated at lower values, with fewer high-value bills extending to the right.

Effect of Bin Size:

With 5 bins, the distribution looks more generalized and smooth, but finer details are hidden.

With 20 bins, we can see more granular patterns, small peaks, and variation more clearly.

Task 2: Box Plot & Outliers

Code (Box Plot)

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='day', y='total_bill', data=df)
plt.xlabel("Day")
plt.ylabel("Total Bill (USD)")
plt.title("Total Bill Distribution Across Days")
plt.show()
```

Code (IQR Calculation Example for Saturday)

```python
sat_data = df[df['day'] == 'Sat']['total_bill']

Q1 = sat_data.quantile(0.25)
Q3 = sat_data.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

Q1, Q3, IQR, lower_bound, upper_bound
```

Markdown Answer:
Highest Median:
The median total bill appears highest on Saturday.

Outliers:
Saturday shows several high-value outliers.

IQR Verification:
Using the IQR method:

Values above Q3 + 1.5×IQR are considered outliers

The extreme values visible in the box plot fall outside this range, confirming they are true outliers

Task 3: Interactive Plot (Plotly)

```python
import plotly.express as px

fig = px.scatter(
    df,
    x="total_bill",
    y="tip",
    color="time",
    hover_data=["day", "size"],
    title="Total Bill vs Tip (Colored by Time)"
)

fig.show()
```

Markdown Answer:

Relationship Between total_bill and tip:
There is a positive relationship — as the total bill increases, the tip amount also tends to increase.

Lunch vs Dinner Patterns:

Dinner transactions generally have higher total bills and tips

Lunch points are more clustered at lower values

This suggests higher spending during dinner

Interesting Data Point:
One noticeable point is a very high total bill with a relatively low tip, which may indicate:

Poor tipping behavior

Service dissatisfaction

Or a large group splitting tip unevenly
