Task 1 — Fetch and Validate API Response

```python
import requests

url = "https://restcountries.com/v3.1/all?fields=name,population,region,area"

try:
    response = requests.get(url)
    print("Status Code:", response.status_code)
    
    if response.status_code == 200:
        data = response.json()
        print("Total countries retrieved:", len(data))
    else:
        print("Failed to fetch data. Please check the API.")
        
except Exception as e:
    print("Error occurred:", str(e))
```

Task 2 — Extract and Create DataFrame

```python
import pandas as pd

# Extract required fields
country_list = []

for country in data:
    name = country.get("name", {}).get("common", None)
    population = country.get("population", None)
    region = country.get("region", None)
    area = country.get("area", None)
    
    country_list.append({
        "common_name": name,
        "population": population,
        "region": region,
        "area": area
    })

df = pd.DataFrame(country_list)

# Display first 10 rows
df.head(10)
```

Task 3 — Summary Analysis

```python
# Region with highest total population
region_population = df.groupby("region")["population"].sum()
highest_region = region_population.idxmax()
highest_population = region_population.max()

# Country with largest area
largest_country = df.loc[df["area"].idxmax(), "common_name"]
largest_area = df["area"].max()

print("Region with highest total population:", highest_region)
print("Total population:", highest_population)

print("\nCountry with largest area:", largest_country)
print("Area:", largest_area)
```
